# 01 — Começando com ficha-py

`ficha-py` é a camada analítica compartilhada do FICHA (ADR 0017): as mesmas
expressões Ibis usadas pelo ETL para escrever os parquets ficam disponíveis
aqui para consulta direta contra o snapshot publicado no Internet Archive.

Nenhum download completo é necessário — o DuckDB `httpfs` faz HTTP range
requests, buscando só os row groups tocados pela query.

```python
# No Colab, rode esta célula primeiro:
# !pip install ficha-py
```

```python
import ficha_py

# Conecta no snapshot mais recente conhecido — troque o mês conforme
# web/public/manifest.json ("current") no momento em que você rodar isto.
con = ficha_py.connect_ia(month="2026-04")
con.list_tables()
```

## Uma empresa por CNPJ

```python
ficha_py.cnpjs(con).filter(ficha_py._.cnpj == "00000001000191").execute()
```

## Sócios e filiais de uma raiz

`socios_de` e `filiais_de` são exatamente as mesmas expressões Ibis usadas
pelo ETL (`ficha_py.views`) — filtram por `cnpj_base` (8 dígitos).

```python
ficha_py.socios_de(con, "00000001").execute()
```

```python
ficha_py.filiais_de(con, "00000001").execute()
```
